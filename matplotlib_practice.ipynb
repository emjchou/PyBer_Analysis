```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot
plt.plot(x_axis, y_axis)
```

```python
#create the plot w ax.plt() (this is object-oriented interface)

#create the figure. this creates the figure and the axis labels
#fig, ax = plt.subplots()
fig=plt.figure()
ax=fig.add_subplot() 

#define axes. access the ax (axis) object and plot data to x and y axis
ax.plot(x_axis, y_axis)
```

```python
#create the plot w ax.plt()
ax=plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#creat the plot and add a label for the legend
plt.plot(x_axis, y_axis, label='Boston', marker="*", color="blue", linewidth=2)

#create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

#set the y limit between 0 and 45
plt.ylim(0, 45)

#create a title
plt.title("Pyber Fare by Month")

#add a grid
plt.grid()

#add legend
plt.legend()
```

```python
fig, ax=plt.subplots()
ax.plot(x_axis, y_axis, label="Boston", color="green", marker="D", linewidth=2)
ax.legend()
ax.set_title("Pyber Fare by Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_ylim(0, 45)
ax.grid()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
#plot
plt.scatter(y_axis, x_axis, color="Red", label="Chicago")

#title and axes labels
plt.title("Pyber Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")

#add limit
plt.xlim(0, 50)

#invert date
plt.gca().invert_yaxis()
```

```python

plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
fig, ax=plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
#create bubble chart
fig, ax=plt.subplots()
ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis], color="skyblue", alpha=0.8, edgecolor="black", linewidth=2, label="Boston")
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_xlim(0, 50)
ax.invert_yaxis()
```

```python
#pie chart, matlab method
plt.pie(y_axis, labels=x_axis)
plt.show()

```

```python
#explode a wedge and add percentages
explode_values=(0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct="%.1f%%")
plt.show()
```

```python
#assign 12 colors, one for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

#assign explode values so that july is exploded
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

#change display size for the figure
plt.subplots(figsize=(8, 8))

#plot the pie chart
plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct="%.1f%%")

plt.show()
```

```python
fig, ax=plt.subplots(figsize=(8, 8))
explode_values=(0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis, labels=x_axis, autopct="%.1f%%", explode=explode_values, shadow=True, startangle=90, counterclock=False, colors=colors)
plt.show()
```
